```python
import keras
import tensorflow as tf
from keras.layers import Dense,Dropout
from keras.models import Sequential
from keras.optimizers import Adam,RMSprop,SGD
from tensorflow.keras import layers, models
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train
```

```python
x_train= keras.utils.to_categorical(x_train)
x_test = keras.utils.to_categorical(x_test)
y_train= keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
x_train= x_train.reshape(16000,256)
x_test = x_test.reshape(4000,256)
x_train = x_train.astype('float')
x_test = x_test.astype('float')
x_train/=255
x_test/=255
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# intialize epochs,batch_size,n_classes
epochs=4
batch_size=4000
n_classes=26
```

# Model Building

```python
model=Sequential()
model.add(Dense(181,activation='relu',input_shape=(256,)))
model.add(Dense(181,activation='relu'))
model.add(Dense(n_classes,activation='sigmoid'))
model.summary()       
```

```python
256*2/3+10
```

```python
from keras.callbacks import EarlyStopping
```

```python
calls=EarlyStopping(monitor='val_loss',patience=2)
```

# Model Compile

```python
model.compile(optimizer=Adam(),loss='categorical_crossentropy',metrics=['accuracy'])
```

# Model Vlidation

```python
history=model.fit(x_train,y_train,epochs,batch_size,validation_data=(x_test,y_test),callbacks=calls)
```

```python
plt.plot(range(len(history.history['accuracy'])), history.history['accuracy'], label='Train')
plt.plot(range(len(history.history['val_accuracy'])), history.history['val_accuracy'], label='Test')
plt.legend(loc='best')
plt.xlabel('No of Epochs')
plt.ylabel('Acc vs Val_Acc')
plt.title('Model Accuracy')
plt.show()
```

```python
plt.plot(range(len(history.history['accuracy'])), history.history['accuracy'], label='Train')
plt.plot(range(len(history.history['val_accuracy'])), history.history['val_loss'], label='Test')
plt.legend(loc='best')
plt.xlabel('No of Epochs')
plt.ylabel('Loss vs Val_Loss')
plt.title('Model Accuracy')
plt.show()
```

```python
val_loss, val_accuracy = model.evaluate(x_test, y_test)
print(f"Validation Accuracy: {val_accuracy:.4f}")
```
