```python
# find mean to replace missing values in Age column
mean=df_train['Age'].mean()
mean
```

```python
df_train
```

```python
# identify outliers
```

```python

Q1=df_train['Age'].quantile(0.25)
Q3=df_train['Age'].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
lower_extreme=Q1-1.5*IQR
upper_extreme=Q3+1.5*IQR
print(lower_extreme,upper_extreme)
```

```python
# Remove outliers
df_train=df_train[(df_train['Age']>=lower_extreme) | (df_train['Age']<=upper_extreme)]
df_train
```

```python
# split the data into training and testing
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=100)
```

```python
import lightgbm as lgb
```

```python
lgb= lgb.LGBMClassifier(n_estimators=100, max_depth=9,max_bins=10,boosting_type='goss',top_rate=0.3,other_rate=0.2)
```

```python
lgb.fit(x_train,y_train)
```

```python
y_pred=lgb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Compare the performance metrics of LightGBM and XGBoost models

XGBoost (XGBM) slightly outperforms LightGBM (LGBM) overall, with a higher accuracy (0.70 vs. 0.68) and better precision, especially for Class 1 (0.66 vs. 0.60), indicating fewer false positives. XGBM also has better recall for Class 0 (0.86 vs. 0.81), meaning it captures more true positives in this class. Both models have similar F1-scores for Class 1 (0.52), but XGBMâ€™s higher F1 for Class 0 (0.78 vs. 0.76) gives it an edge. Thus, XGBM shows stronger overall performance, though LGBM may be preferable if prioritizing recall for Class 1.

```python
# Visualize and interpret the results to identify the strengths and weaknesses of each algorithm.
```

```python
import numpy as np

# Performance metrics from classification reports
metrics = ["Precision", "Recall", "F1-Score", "Accuracy"]
xgbm_scores = [0.69, 0.70, 0.69, 0.70]  
lgbm_scores = [0.67, 0.68, 0.67, 0.68]  

# Bar chart for comparison
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x - width/2, lgbm_scores, width, label="LGBM")
ax.bar(x + width/2, xgbm_scores, width, label="XGBM")

ax.set_xlabel("Metrics")
ax.set_title("Comparison of LGBM and XGBM Performance Metrics")
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
plt.show()
```
