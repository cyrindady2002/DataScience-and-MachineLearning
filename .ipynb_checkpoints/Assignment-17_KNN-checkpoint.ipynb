```python
# identify null values
df.isnull().sum()
```

```python
num_col=df[[i for i in df.columns if df[i].dtype!='O']]
num_col
```

```python
# identify outliers
```

```python
Q1=num_col.quantile(0.25)
Q3=num_col.quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
lower_extreme=Q1-1.5*IQR
upper_extreme=Q3+1.5*IQR
print(lower_extreme ,upper_extreme)
```

```python
# Remove outliers in columns

for col in num_col:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
```

```python
# Encode categorical variables using OneHotEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot=OneHotEncoder()
```

```python
df1=pd.DataFrame(one_hot.fit_transform(df[['animal name']]).toarray())
```

```python
df1
```

```python
df=df.join(df1)
```

```python
df
```

```python
df['animal name']
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.drop((['animal name']),axis=1,inplace=True)
```

```python
df=df.rename(columns={11:'a',12:'b',13:'c',14:'d',15:'e',16:'f',16:'g',17:'h',18:'i',19:'j',20:'k'})
```

```python
df
```

```python
target=df[['type']]
target
```

```python
features=df.drop(columns='type')
features
```

```python
# Split the dataset into training and testing sets (80% training, 20% testing).
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=42)
```

```python
# Convert all column names to strings
x_train.columns = x_train.columns.astype(str)
x_test.columns=x_test.columns.astype(str)
```

```python
# KNN
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
var=range(1,20)
```

```python
params=dict(n_neighbors=var)
```

```python
params
```

```python
grid_search=GridSearchCV(knn,params,cv=4)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

## Interview Questions and Answers
What are the key hyperparameters in KNN?

K (number of neighbors): Defines the number of nearest neighbors to consider when making a classification decision.
Distance metric: Determines how the distance between points is calculated, affecting which neighbors are selected.
Weights: Defines if all neighbors should be weighted equally or if closer neighbors should have more influence on the classification.

What distance metrics can be used in KNN?

Euclidean distance: The most common and default metric; it calculates the straight-line distance between points.
Manhattan distance: Measures distance by only moving along grid lines, calculating the sum of absolute differences.
Minkowski distance: Generalizes both Euclidean and Manhattan distances, parameterized by a power factor p.
